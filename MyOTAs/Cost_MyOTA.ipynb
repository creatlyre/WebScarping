```python
import math

LICENCE_PRO_MULTIPLIER = 3 # multiplyaer for the addtional categories added to license
LICENCE_PREMIUM_MULTIPLIER = 1
LICENCE_STARTER_MULTIPLIER = 0
POLAND_TAX = 1.23 
STORAGE_COST = 1.5 # fixed cost of storage on azure data stroage in usd
DAYS_IN_MONTH=30
RED_BOLD_START = "\033[1;31m"  # Red and bold start
COLOR_END = "\033[0m"          # Reset color and style
BOLD_START = "\033[1m"  # Start bold text
BOLD_END = "\033[0m" # end bold text
AVG_PAGE_GLOBAL = 95 # based on the 17 citires which we were collecting in the past (Viator)
AVG_PAGE_OTHER = 28 # based on the 2 citires which we were collecting in the past (Viator) Lisbon and Porto
    
    
def backend_support(price_per_h, cities_count_total, additional_time=0):
# """Back end support is the price per month for hours working on My OTAs requried on daily basis which depends 
#    on the amount of cities which we are collecting """   
    daily_time = cities_count_total/100
    cost = (daily_time * DAYS_IN_MONTH)*price_per_h + additional_time*price_per_h
    return cost

def count_credits(license_type, cities_count, api_cost=49,  api_limits = 100000):
# """ 
# Script to calculte what would be the price of SCraper and Storage for number of cicites and license type
# """

    if license_type.lower() == 'premium':
        license_multiplier = LICENCE_PREMIUM_MULTIPLIER
        threshold = 0.35
    elif license_type.lower() == 'pro':
        license_multiplier = LICENCE_PRO_MULTIPLIER
        threshold = 0.5
    elif  license_type.lower() == 'starter':
        license_multiplier = LICENCE_STARTER_MULTIPLIER
        threshold = 0.25
    else:
        raise ValueError("license_type must be 'premium' or 'pro'")

    count_global_category = cities_count
    count_other_category = cities_count * license_multiplier
    credits_per_day = (AVG_PAGE_GLOBAL * math.ceil(count_global_category * threshold)) + \
                      (AVG_PAGE_OTHER * math.ceil(count_other_category * threshold))
    credits_per_month = credits_per_day * DAYS_IN_MONTH
    cost_per_month_storage = round(STORAGE_COST * cities_count, 2)
    cost_per_month_api = round((credits_per_month/api_limits) * (api_cost*POLAND_TAX), 2)
    cost_per_month = round((credits_per_month/api_limits * api_cost)*POLAND_TAX + STORAGE_COST * cities_count, 2)
    
    
    print(f'For the {BOLD_START}{license_type.upper()}{BOLD_END} licence with {cities_count} cities at threshold {threshold*100}%. \
    \nWe would need {credits_per_month} credits monthly costing {RED_BOLD_START}{cost_per_month}${COLOR_END} \
    \nSCRAPER: {cost_per_month_api}$ STORAGE: {cost_per_month_storage}$\n') 
    return cost_per_month
def new_city_added(price_per_h, hours):
    cost = hours * price_per_h
    return cost
    
total_amount_of_cities_in_app = 45
backend_support_cost = backend_support(35, total_amount_of_cities_in_app, additional_time=5)
new_city_added_cost = new_city_added(35, 7)
print(f'{RED_BOLD_START}VIATOR PRICES{COLOR_END}\n')
count_credits(license_type='starter', cities_count=10)   # Calculate for 5 cities with STARTER licence
count_credits(license_type='premium', cities_count=10)   # Calculate for 5 cities with PREMIUM licence
count_credits(license_type='pro'    , cities_count=10)   # Calculate for 5 cities with PRO licence
print(f'Support cost for month: {RED_BOLD_START}{backend_support_cost}${COLOR_END}\
\nNew city price: {RED_BOLD_START}{new_city_added_cost}${COLOR_END}')


print(f'\n\n{RED_BOLD_START}With higher API Limit{COLOR_END}\n\n')
count_credits(license_type='starter', cities_count=5, api_cost=150, api_limits=1000000)   # Calculate for 5 cities with STARTER licence
count_credits(license_type='premium', cities_count=5, api_cost=150, api_limits=1000000)   # Calculate for 5 cities with PREMIUM licence
count_credits(license_type='pro'    , cities_count=5, api_cost=150, api_limits=1000000)   # Calculate for 5 cities with PRO licence


    
print(f'{RED_BOLD_START}Backlog to pay: 81512 EUR xD{COLOR_END}')

```
