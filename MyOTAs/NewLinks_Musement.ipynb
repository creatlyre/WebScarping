```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
import pandas as pd
import time

list_links = []
name_links = []
raw_links = []
what_is_it = []
append_city = []
```

```python
city_list = [
"Amsterdam",
"Athens",
"Barcelona",
"Berlin",
"Cancun",
"Dubai",
"Dublin",
"Dubrovnik",
"Edinburgh",
"Florence",
"Istanbul",
"Krakow",
"Las-Vegas",
"Lisbon",
"London",
"Madrid",
"Milan",
"Naples",
"New-York-City",
"Paris",
"Porto",
"Rome",
"Palermo",
"Venice",
"Taormina",
"Capri",
"Sorrento",
"Mount-Etna",
"Mount-Vesuvius",
"Herculaneum",
"Amalfi-Coast",
"Pompeii"]
```

```python
city_list = ['Amsterdam']
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()
for new_city in city_list:
    base_url = f"https://www.musement.com/uk/{new_city}/?sort_by=-relevance-city&limit=12"
    driver.get(base_url)
#     city_input = driver.find_elements(By.CLASS_NAME, 'src-carouselContainer-3ykt')
#     time.sleep(1)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    links = soup.find_all(attrs={'data-test-id': 'category-button'})
    list_links.append(base_url.lower())
    name_links.append('Global')
    raw_links.append('Global')
    what_is_it.append('Category')

    append_city.append(new_city)

    for link in links:
        name_category = '&category_in=new-' + link.text.strip().replace(' ', '-').replace('-&', '').replace('-and', '').lower()
        list_links.append(base_url.lower()+name_category)
        name_links.append(link.text.strip())
        raw_links.append(link.text.strip())
        what_is_it.append('Category')
        append_city.append(new_city)

    attractions = soup.find_all(attrs={'class': 'attraction_card_content'})
    for attraction in attractions:
        list_links.append('https://www.musement.com'+attraction.find('a')['href'])
        name_links.append(attraction.find(attrs={'data-test': 'AttractionCard__title'}).text.strip())
        raw_links.append(attraction.find(attrs={'data-test': 'AttractionCard__title'}).text.strip())
        what_is_it.append('Attraction')
        append_city.append(new_city)
df_links = pd.DataFrame({'URL': list_links, 'Category':name_links, 'RawCategory': raw_links,'WhatIsIt': what_is_it, 'City': append_city})
df_links.drop_duplicates(subset=('URL'), inplace=True)
# driver.quit()
df_links
```
