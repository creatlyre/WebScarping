```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
import pandas as pd
import time

list_links = []
name_links = []
raw_links = []
append_city = []
```

```python
city_list = [
"Amsterdam",
"Athens",
"Barcelona",
"Berlin",
"Cancun",
"Dubai",
"Dublin",
"Dubrovnik",
"Edinburgh",
"Florence",
"Istanbul",
"Krakow",
"Las-Vegas",
"Lisbon",
"London",
"Madrid",
"Milan",
"Naples",
"New-York-City",
"Paris",
"Porto",
"Rome",
"Palermo",
"Venice",
"Taormina",
"Capri",
"Sorrento",
"Mount-Etna",
"Mount-Vesuvius",
"Herculaneum",
"Amalfi-Coast",
"Pompeii"]
```

```python
city_list = ['Capri']
```

```python

# new_city = 'Porto'
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()
base_url = "https://www.viator.com"
for new_city in city_list:
    driver.get(base_url)
    city_input = driver.find_element(By.ID, 'react-select-homepageTypeaheadSearch-input')
    city_input.send_keys(new_city)
    button = driver.find_element(By.CSS_SELECTOR, '[data-automation="experiences-search-button"]')
    # click the button
    button.click()
    time.sleep(3)
    current_url = driver.current_url
    print(current_url)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    links = soup.find_all('div', {'class': 'dynamic-filter slick-slide'} )
    list_links.append(current_url)
    name_links.append('Global')
    raw_links.append('/Global/Global')
    append_city.append(new_city)

    for link in links:
        list_links.append(base_url+link.find('a')['href'])
        name_links.append(link.text.strip())
        raw_links.append(link.find('a')['href'])
        append_city.append(new_city)

    left_box = soup.select("[class*=tour-card]") 
    for item in left_box:
        details = item.find('div', {'class': 'media'}).find('a')
        list_links.append(base_url + details['href'])
        name_links.append(details.text)
        raw_links.append(details['href'])
        append_city.append(new_city)

    
df_links = pd.DataFrame({'URL': list_links, 'Category':name_links, 'RawCategory': raw_links, 'City': append_city})
df_links.drop_duplicates(subset=('URL'), inplace=True)
driver.quit()
df_links['MatchCategory'] = df_links['RawCategory'].map(lambda x: x.split('/')[2])
# df_links['URL'] = df_links['URL'].str.replace('https://', '')
df_links
```
